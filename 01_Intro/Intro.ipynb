## Data Types and Operators

### Lists

```
# Creating lists
x = list()
y = []
x = list((2, 1, 4, 8))
y = [2, 1, 4, 10, 20, 40]
print('About x:')
print('type(x) - ', type(x))
print('x       - ', x)
print('len(x)  - ',len(x))
print(' ')
print('About y:')
print('type(y) - ', type(y))
print('y       - ', y)
print('len(y)  - ',len(y))
```

```
# Indexing lists
x_1 = x[2]
# Index slicing
y_1 = y[2:]
y_2 = y[2:4]
print('x[2]  - ', x_1)
print('y_1 (y[2:])  - ', y_1)
print('y_2 (y[2:4]) - ', y_2)
```

```
# Append and Extend
x.append(y)
print(x)
y.extend(y_1)
print(y)
```

## Classes

```
class Model(object):
    # class initialization and attributes
    def __init__(self, params, model_type, name='Neural Network'):
        self.name = name
        self.params = params
        self.model_type = model_type

    # internal functions denoted by __.__
    # str
    def __str__(self):
        return 'Readable string : This is a '+ self.model_type + ' ' + self.name + ' with ' + str(self.params) + ' params'

    # representation
    def __repr__(self):
        return self.model_type + ' ' + self.name + ' with ' + str(self.params) + ' params'

    # method
    def printParams(self, multiplier):
        print(self.params*multiplier)
```

```
m1 = Model(5, 'Supervised', 'LinearRegression')
```

```
m1
```

```
print(m1)
```

```
m1.name
```

```
m1.printParams(10)
```

```
# Try changing kernel size and observe what happens
kernel = np.ones((10,10), np.float32)/100

# Arguments are (source_image, depth, kernel)
dst = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)

plt.subplot(121)
plt.imshow(img[:,:257,:]),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(dst[:,:257,:]),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

### Histogram Equalization

```
img = cv2.imread('/content/ME592_S24/01Intro/lung.PNG', 0)
print(img)
print(img.dtype)
print(img.shape)
cv2_imshow(img)
```

```
# Arguments are (img, bins, range)
_ = plt.hist(img.flatten(),256,[0,256], color = 'r')
```

```
# Histogram equalization can be used to improve constrast of img
equ = cv2.equalizeHist(img)
_ = plt.hist(equ.flatten(),256,[0,256], color = 'r')
```

```
cv2_imshow(img)
cv2_imshow(equ)
```

```
threshold_value = 150

# Apply the threshold
_, thresholded_image = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)

# Display the thresholded image
cv2_imshow(thresholded_image)
```

### Denoising

```
# Median filter is one way of denoising
img = cv2.imread('/content/ME592_S24/01Intro/Elephant-1.png')[:,:257,:]
median = cv2.medianBlur(img, 3)
cv2_imshow(img)
cv2_imshow((cv2.Canny(img,100,200)))
cv2_imshow(median)
edges = cv2.Canny(median,100,200)
cv2_imshow(edges)
```

```
import scipy.signal
# can be used to compue moving average
def apply_convolution(sig, window):
    conv = np.repeat([0., 1., 0.], window)
    print('Convolution Kernel:', conv)
    filtered = scipy.signal.convolve(sig, conv, mode='same') / window
    return filtered

# apply a median filter, arguments are (volume, kernel)
filtered = scipy.signal.medfilt(timedata_np[:,1], 7)
# apply a convolution filter
convolved = apply_convolution(timedata_np[:,1], 20)
plt.plot(filtered)
plt.show()
plt.plot(convolved)
plt.show()
```

## Dimensionality Reduction

```
from keras.datasets.mnist import  load_data
_, (images, labels) = load_data()
print(images.shape, labels.shape)
```

```
idx = np.random.randint(0, len(images))
plt.imshow(images[idx])
print(labels[idx])
plt.show()
```

```
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```
pca = PCA(n_components=2)
pca_result = pca.fit_transform(images.reshape(10000,28*28))
pca_result.shape
```

```
plt.figure()
plt.scatter(pca_result[:,0], pca_result[:,1], c=labels)
plt.colorbar()
plt.show()
```

```
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
import time
start = time.time()
tsne_results = tsne.fit_transform(images.reshape(10000,28*28))
end = time.time()
print('Time', end-start)
```

```
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=labels)
plt.colorbar()
plt.show()
```

```
# applying PCA n high-dimensional data, then using t-SNE can speed up process
pca = PCA(n_components=50)
pca_result_int = pca.fit_transform(images.reshape(10000,28*28))
start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(pca_result_int)
end = time.time()
print('Time', end-start)
```

```
plt.figure()
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=labels)
plt.colorbar()
plt.show()
```
